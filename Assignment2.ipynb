```python
final["Speed Group"]=final["Speed Limit"]
```

```python
for i in range(len(final)):   
    if final["Speed Group"].iloc[i] in range(0,31):
        final["Speed Group"].replace(final["Speed Group"].iloc[i],"10_30", inplace=True)
    elif final["Speed Group"].iloc[i] in range(31,61):
        final["Speed Group"].replace(final["Speed Group"].iloc[i],"40_60", inplace=True)
    elif final["Speed Group"].iloc[i] in range(61,91):
        final["Speed Group"].replace(final["Speed Group"].iloc[i],"70_90", inplace=True)
    elif final["Speed Group"].iloc[i] in range(91,121):
        final["Speed Group"].replace(final["Speed Group"].iloc[i],"100_120", inplace=True)
    elif final["Speed Group"].iloc[i] in range(121,151):
        final["Speed Group"].replace(final["Speed Group"].iloc[i],"120_150", inplace=True)
    
```

```python
final
```

```python
final["Day_Time"]=final["Age"]
```

```python
for i in range(len(final)):   
    if final["Day of week"].iloc[i]=="Weekday" and final["Time of day"].iloc[i]=="Night":
        final.loc[i,"Day_Time"]="weekday_night"
    elif final["Day of week"].iloc[i]=="Weekday" and final["Time of day"].iloc[i]=="Day":
        final.loc[i,"Day_Time"]="weekday_day"
    elif final["Day of week"].iloc[i]=="Weekend" and final["Time of day"].iloc[i]=="Night":
        final.loc[i,"Day_Time"]="weekend_night"
    elif final["Day of week"].iloc[i]=="Weekend" and final["Time of day"].iloc[i]=="Day":
        final.loc[i,"Day_Time"]="weekend_day"
```
